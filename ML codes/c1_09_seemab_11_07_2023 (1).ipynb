```python


import pandas as pd
df=pd.read_csv('iris.csv')
df
```

```python
# feature and label separation

x=df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
print(x)
y=df['species']
print(y)
```

```python
#  2-      Model training of training data set

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

x_train,x_test,y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=23)

model=DecisionTreeClassifier()
model.fit(x_train,y_train)

```

```python
# 3-      Model prediction of set data set

y_prediction=model.predict(x_test)
y_prediction



y_pred_df=pd.DataFrame({"y_test":y_test,"y_pred":y_prediction})
y_pred_df
```

```python
#  4-      Plotting the tree at large

import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

plot_tree(model,filled=True);
#plt.figure(figsize=(40,40))
```

```python
#  Calculate the Entropy

import math

total = len(y)
class_counts = y.value_counts()
print(class_counts)

entropy = 0.0
for i in class_counts:
    print(i)
    probability = i / total
    #print(probability)
    entropy -= probability * math.log2(probability)

print('Entropy:', entropy)
```

```python
#  Calculate Gini Formula

import numpy as np

def gini_index(labels):
    classes, count = np.unique(labels, return_counts=True)

    prob = count / len(labels)
    gini = 1 - np.sum(prob ** 2)
    return gini



gini = gini_index(y)
print("Gini Index:", gini)
```
