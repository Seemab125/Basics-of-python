```python
import os
import cv2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
print(os.listdir(r'C:\Users\HP\Desktop\Image processing Ml\fruits-360-original-size\fruits-360-original-size\Training'))
```

```python
dataset = []
```

```python
folder_paths = [
    r"C:\Users\HP\Desktop\Image processing Ml\fruits-360-original-size\fruits-360-original-size\Training\apple_6",
    r"C:\Users\HP\Desktop\Image processing Ml\fruits-360-original-size\fruits-360-original-size\Training\apple_braeburn_1",
]
```

```python
for i in folder_paths:
    folder_name = os.path.basename(i)
    
    for file_name in os.listdir(i):
        image_path=os.path.join(i, file_name)
        
        if os.path.isfile(image_path): # Only consider files
            # Load the image using OpenCV
            image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
            
            if image is not None:
                # Resize the grayscale image to 250X250 pixels
                resized_image = cv2.resize(image, (100, 100))
                
                # Flatten the image and append each pixel as a separate feature al
                flattened_image = resized_image.flatten().tolist()
                dataset.append(flattened_image + [folder_name])
```

```python
"""Convert the dataset to a pandas DataFrame"""
df = pd.DataFrame(dataset)
"""Print the DataFrame"""
df
```

```python
last_label_column=df.iloc[:,-1]
last_label_column
```

```python
# Count the occurrences of each unique label

label_counts=last_label_column.value_counts()
label_counts
```

```python
# creating a pie chart

plt.pie(label_counts, labels=label_counts.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Distribution of Labels')
plt.show()
```

```python
# Create a bar chart

plt.bar(label_counts.index , label_counts.values)
plt.xlabel('labels')
plt.ylabel('count')
plt.title('Distribution of Labels')
plt.show
```

```python
#  "Normalize the pixel values between 0 and 1"

X=df.iloc[:,:10000]
X=X/255
```

```python
X
```

```python
Y=df.iloc[:,-1]
Y
```

```python
from sklearn.preprocessing import LabelEncoder

y_encoded=LabelEncoder()
yy=y_encoded.fit_transform(Y)
yy
```

```python
y_series=pd.Series(yy)
y_series
```

```python
df=pd.concat([X, y_series], axis=1)
df
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test, y_train,y_test=train_test_split(X,y_series,test_size=0.2,random_state=42)
```

```python
from sklearn.svm import SVC
svm_classifier=SVC(kernel='poly',C=1.0)
svm_classifier.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Predict the labels for the test set
y_pred = svm_classifier.predict(X_test)

# Calculate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Print classification report
print(classification_report(y_test, y_pred))

# Print confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(conf_matrix)
```
