# task 1

```python
import pandas as pd

# Read the "Orders" sheet from the Excel file
df = pd.read_excel('Sample_Superstore.xls', sheet_name='Orders')

# Remove duplicate columns
df = df.loc[:, ~df.columns.duplicated()]

# Remove duplicate rows
df = df.drop_duplicates()

# Remove blank columns
df = df.dropna(axis=1, how='all')

# Remove blank rows
df = df.dropna(axis=0, how='all')

# Print the cleaned DataFrame
df
```

# Task 2

```python
import pandas as pd

# Assuming 'df' is your DataFrame containing the relevant columns

df['Profit'] = pd.to_numeric(df['Profit'], errors='coerce').fillna(0)
profit = df.groupby("Order Date")["Profit"].sum()

max_profit = profit.max()
date_of_max_profit = profit.idxmax()

print('Date:', date_of_max_profit)
print('Max Profit:', max_profit)
```

# task 3

```python
#    task 3
#   Task 3: Calculate average sales w.r.t. States.
    
import pandas as pd

average_sales = df.groupby("State")["Sales"].mean()

print(average_sales)
```

#   task 4

```python
    # Task 4: On which Order Date, the Super Store Company generated lowest profit
    #  and display the customer name and ID.
    
    
import pandas as pd

df['Profit'] = pd.to_numeric(df['Profit'], errors='coerce').fillna(0)
lowest_profit = df.groupby("Order Date")["Profit"].sum().idxmin()

lowest_profit_orders = df[df["Order Date"] == lowest_profit]
customer_info = lowest_profit_orders[["Customer Name", "Customer ID"]]

print("Order Date with Lowest Profit:", lowest_profit)
print("Customer Information:")
print(customer_info)
```

# task 5

```python
#   Task 5:    Display the most used Ship Mode.

import pandas as pd
ship_mode = df['Ship Mode'].value_counts().idxmax()
print('Most Used Ship Mode:', ship_mode)
```

# task 6

```python
#  task 6: display arizonas customer name who gave highest profit to company


import pandas as pf
arizona_customers = df[df['State'] == 'Arizona']
h_p_name = arizona_customers.loc[arizona_customers['Profit'].idxmax(), 'Customer Name']
print("Name of customer with Highest Profit in Arizona:", h_p_name)
```

# task 7

```python
#  Task 7: Display the name of the state that generated the most profits.

import pandas as pd
state_profits = df.groupby("State")["Profit"].sum()
most_profitable_state = state_profits.idxmax()
print("THe state that generated the most profits : ", most_profitable_state )
```
