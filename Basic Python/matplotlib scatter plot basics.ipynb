```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[10,12,15,18,20]

plt.scatter(x,y)
plt.xlabel("X-AXIS")
plt.ylabel("Y-AXIS")
plt.grid(axis='y',color='hotpink')
plt.grid(axis='x',color='hotpink')
plt.show()

```

```python
x1=[1,2,3,4]
y1=[10,20,30,40]

x2=[3,4,5,6]
y2=[5,10,15,20]
plt.plot(x1,y1)
plt.plot(x2,y2)

```

```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([1, 5 ]) 
y1 = np.array([10, 20])

x2 = np.array([3, 7, 4])  
y2 = np.array([5, 21, 5])

plt.plot(x1, y1, label='Line 1')  # Plot for Line 1
plt.plot(x2[:len(y2)], y2)  # Plot for Line 2, using the same number of data points as y2
plt.xlabel("X-AXIS")
plt.ylabel("Y-AXIS")
plt.grid(axis='y')  # Show grid lines along the y-axis
plt.show()  # Display the plot
```
